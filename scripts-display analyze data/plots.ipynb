```python
def plot_Pauli_error_rate(fidelity_list,nqubit,pauli_request_list,figure):
    """
    plot Pauli error rate for the fidelity_list which is the fidelity list directly imported from files in the notebook 
    plot this trace in figure
    """
    dic_temp={}
    for pauli in pauli_request_list:
        dic_temp[pauli[::-1]] = mean(fidelity_list[pauli])

    error_list = fidelity_to_error(dic_temp,nqubit)
    pauli_request_list_temp=pauli_request_list.copy()
    pauli_request_list_temp.remove("II")
    # plt.figure(figsize=(9, 6))
    ax= figure.add_axes((1,1,1,1))
    ax.bar([pauli_label for pauli_label in pauli_request_list_temp],[round(error_list[pauli_label],7) for pauli_label in pauli_request_list_temp])
```
