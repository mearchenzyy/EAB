```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample ordinal x-values
x_values = ['A', 'B', 'C', 'D', 'E']

# Generate sample datasets
dataset1 = np.array([10, 20, 15, 25, 30])
dataset2 = np.array([15, 25, 18, 28, 35])

# Add some random noise to the x-values
noise = 0.4 # Adjust the noise level as needed
x_values_offset = np.arange(len(x_values)) + np.random.uniform(-noise, noise, len(x_values))

# Plot the datasets with the offset x-values
plt.plot(x_values_offset, dataset1, marker='o', label='Dataset 1')
plt.plot(x_values_offset, dataset2, marker='x', label='Dataset 2')

# Set x-axis tick labels to original ordinal values
plt.xticks(np.arange(len(x_values)), x_values)

# Add legend and labels
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Datasets with Offset X-values')

# Show plot
plt.show()
```
